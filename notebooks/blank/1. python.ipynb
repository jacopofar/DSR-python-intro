# 15. List Range Command

The `range()` function in Python is used to generate a sequence of numbers.

This is a lazy function (aka "generator"), it does generate them "on demand"

````python
for i in range(100_000_000):
    if i == 9000:
        print(i)
        break
````

````python
a = [1, 2, 3]
b = a[:]
print(a)
print(b)
a[1] = 9000
print(a)
print(b)
c = [1,2,3,4,5]
print(c[1:4:2])
````
